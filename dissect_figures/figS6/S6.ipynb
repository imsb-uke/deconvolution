```python
import os
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datadir = "data"
savedir = "plots"

if not os.path.exists(savedir):
    os.mkdir(savedir)
```

```python
df = pd.read_csv(os.path.join(datadir, "ds_results_r.csv"), index_col=0)
```

```python
df = df[[col for col in df.columns if "8k test" not in col]]
```

```python
melted = df.reset_index().melt(id_vars="index")
```

```python
sns.set(style="ticks", font_scale=1.2)
plt.figure(figsize=(10,5))
ax=sns.barplot(data=melted, x="variable",  y="value", hue="index")

ax.set_ylim()
plt.legend(bbox_to_anchor=(1,1), frameon=False)
plt.ylabel("r")
plt.xlabel("")
plt.savefig(os.path.join(savedir, "domain_shift.pdf"), bbox_inches="tight")
plt.savefig(os.path.join(savedir, "domain_shift.png"), bbox_inches="tight")
```
