```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from anndata import AnnData
import scanpy as sc
```

```python
data = sc.read("data/st_lymph.h5ad")
```

```python
gc_annots = pd.read_csv("data/manual_GC_annot.csv", index_col=0)
```

```python
data.obs["GC_spot"] = gc_annots.cell_type.tolist()
```

```python
cmap="magma"
```

```python
sc.set_figure_params(dpi=250)
sns.set(font_scale=1.5)
ax=sc.pl.spatial(data, color="GC_spot", cmap=cmap, colorbar_loc=None,
             title="Ground truth", return_fig=True, show=False)
ax[0].set_xlabel("")
ax[0].set_ylabel("")

plt.savefig("plots/GC.png", bbox_inches="tight")
```
